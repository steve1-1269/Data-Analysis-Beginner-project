```python
#  import modules

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
plt.figure(figsize= (16.4,14.5))
```

```python
#  read data

df= pd.read_csv(r"C:\Users\MINE\Desktop\data science\movies.csv")
df.sample(5)
```

```python
#  find missing values

for col in df.columns:
    percentage= np.mean(df[col].isnull())
    print('{} - {}%'.format(col, percentage))
```

```python
# data types
df.dtypes
```

```python
df['budget']=df['budget'].fillna(0)
df['gross']=df['gross'].fillna(0)
```

```python
#  change datatype of columns
df['budget']= df['budget'].astype('int64')
df['gross']= df['gross'].astype('int64')
```

```python
df.dtypes
```

```python
#  Drop duplicate

df.drop_duplicates()
df.query('genre	== "Drama" ')
```

```python
#  create correct year column
df['released']= df['released'].astype(str)
df[['day','year and country']] = df['released'].str.split('(', expand= True)
```

```python
df=df.sort_values(by= 'gross', inplace= False, ascending= False)
```

```python
#  Drop unwanted column
df.drop('day',axis=1)
```

```python
#  Scatter plot with budget vs gross revenue

plt.scatter(x= df['budget'], y= df.gross)

plt.title('Budget Vs Gross Earning')
plt.xlabel('Budget For Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# plot the budget vs gross using the seaborn


sns.regplot(x='budget', y='gross', data= df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
#  looking at correlatiobn
#  created a new data frame of numbers cause coreelation only works on numbers
df1= df[['year','score','budget','gross','runtime','votes']]
df1.corr( method= 'pearson')
```

```python
df1.corr(method= 'kendall')
```

```python
df1.corr(method= 'spearman')
```

```python
corr_matrix=df1.corr( method= 'pearson')
```

```python
sns.heatmap(corr_matrix, annot= True)
plt.title('Correlation Matrix For Numeric Features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')

plt.tight_layout()
plt.show()
```

```python
df_numerize= df


for col in df_numerize.columns:
    if (df_numerize[col].dtype == 'object'):
        df_numerize[col]=  df_numerize[col].astype('category')
        df_numerize[col]=   df_numerize[col].cat.codes

df_numerize
    
```

```python
corr=df_numerize.corr( method= 'pearson')
```

```python
plt.figure(figsize= (15,10))
sns.heatmap(corr, annot= True)

plt.title('Correlation Matrix For Numeric Features')
plt.xlabel('Movie features')
plt.ylabel('Movie features')

plt.tight_layout()
plt.show()
```
