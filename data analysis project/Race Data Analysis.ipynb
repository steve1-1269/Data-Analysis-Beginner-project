# Clean data

# ONly want usa races, 50km or 50mi, 2020

```python
df1= df
df1=df1[df1['Year of event'] == 2020] 
df1=df1[df1['Event distance/length'].isin(['50km','50mi'])]
df1=df1[df1['Event name'].str.split('(').str.get(1).str.split(')').str.get(0)== 'USA']
df1.head(5)
```

# Correlations

```python
df2= df1

for col in df2.columns:
    if df2[col].dtype == 'object':
        df2[col]= df2[col].astype('category')
        df2[col]= df2[col].cat.codes

df2
```

```python
corr2= df2.corr()
corr2
```

```python
sns.regplot(data= df2, x= 'Age', y='athlete_average_speed')
```

```python
sns.lmplot(data= df2, y='athelete_performance', x='Age', hue= 'athlete_gender')
```

```python
plt.figure(figsize=(20,15))

sns.heatmap(corr2,annot= True)
```
